##  -> We have created an Inventory through the Admin Interface file and displaying that Inventory here to the User
### Below we have called Function "bill" through which the user buys products. Also the INVENTORY gets Updated simultaneously !

```python
import json

# opening the Inventory file and reading to use it in the program
fd = open("Inventory File.json","r")
js = fd.read()
fd.close()

# json.loads() function converts back the records from json string to original python dictionary object
records = json.loads(js)  

# printing the LIVE Inventory
print("--------MENU-----------\n")

for key in records.keys():
    print(key,":",records[key]["name"],"    |",records[key]["price"],"   |",records[key]["qt"])

    
# informing the shopper about the available discounts 
print("\n------Discounts available------\n")
print("10% discount over order value of more than 100  Rs. !")
print("30% discount over order value of more than 1000 Rs. !")


# registering customer details to generate sales report
print("\n------Customer Details------\n")
name    = str(input("Enter your name      :"))
ph_no   = str(input("Enter your phone no. :"))
mail_id = str(input("Enter your mail id   :"))


# taking input about the product the shopper wants to buy
print("\n------Enter Product Details------\n")
ui_pi = str(input("Enter Product ID :"))
ui_qt = int(input("Enter Quantity   :"))


# displaying order details along with final cost to the shopper through "bill" function below
print("\n------Order Details------\n")


# function to make the purchase 
def bill(ui_pi, ui_qt):
    
    # from here checking shopper entered qt to check whether there is sufficient qt available
    # also cheking if the qt entered is valid or not 
    if(records[ui_pi]["qt"] == 0):
        
        print("The Product is out of Stock!")
        
    else:     
        
        # this block runs when the qt entered is valid and there is sufficient qt available in the Inventory
        if(ui_qt > 0 and ui_qt <= records[ui_pi]["qt"]):
            
            print("Product Name        :",records[ui_pi]["name"])
            print("Product Price (Rs.) :",records[ui_pi]["price"],"Rs")
            print("Quantity Available  :",records[ui_pi]["qt"])

            print("\n----------BILL---------\n")
            
            amt=records[ui_pi]["price"]*ui_qt
            
            if(amt > 100):        
                
                print("Amount Payable before dsicount :",amt,"Rs")
                amt = 9/10 * (amt)
                print("Amount Payable after discount of 10% :",amt,"Rs")  
                
            elif(amt > 1000):
                
                print("Amount Payable before discount :",amt,"Rs")
                amt = 7/10 * (amt)
                print("Amount Payable after discount of 30% :",amt,"Rs")
                
            else:
                
                print("Amount Payable :",amt,"Rs")

            records[ui_pi]["qt"] = records[ui_pi]["qt"] - (ui_qt)

            print("-----------------------\n")

            js = json.dumps(records)  # converting records object into a json string and assigning it to "js" variable 

            fd = open("Inventory File.json","w")
            fd.write(js)
            fd.close()
            
            ss = ui_pi+","+name+","+ph_no+","+mail_id+","+records[ui_pi]["name"]+","+str(ui_qt)+","+str(amt)+"\n"
            s =  open("Sales Report.txt","a")
            s.write(ss)
            s.close()
            
            print("Inventory updated !")
            print("Sales Report file updated !")
            
        else: 
            
            if(ui_qt > records[ui_pi]["qt"]):
                
                print("We only have",records[ui_pi]["qt"],"quantity!.\nPlease reduce your quantity.")
                
                ui_qt = int(input("Enter Quantity   :"))
                
                bill(ui_pi, ui_qt) 
                
            else:
                
                print("Invalid Entry!\nPlease enter again.")
                
                ui_qt = int(input("Enter Quantity   :"))
                
                bill(ui_pi, ui_qt) 
                

bill(ui_pi, ui_qt)
```
